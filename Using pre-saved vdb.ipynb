## Using pre-existing vector data base for RAG

### Redefine the persistence directory and the embedding function

```python
from langchain.vectorstores import Chroma

PERSIST_DIR = "./bns2"

from langchain_aws import BedrockEmbeddings
embedding_model = BedrockEmbeddings(model_id="amazon.titan-embed-text-v1")
text = "Example text for understanding text-embeddings"
query_result = embedding_model.embed_query(text)
query_result[:5]
```

###  2. Load the existing vector store from the directory
####  Chroma will read the saved data and collections from this directory

```python

loaded_vectorstore = Chroma(
    collection_name="bns2",
    persist_directory=PERSIST_DIR, 
    embedding_function=embedding_model
)

print(f"Chroma DB successfully loaded from {PERSIST_DIR}")

```

### 3. Perform a similarity search

```python

query = "criminal activity"
k_results = 3 # Number of nearest neighbors to retrieve

# similarity_search will automatically use the loaded embedding function 
# to embed your query and find similar documents
docs = loaded_vectorstore.similarity_search(query, k=k_results)

```

### 4. Print the results

```python

print("\n--- Similarity Search Results ---")
for i, doc in enumerate(docs):
    print(f"Result {i+1}:\n{doc.page_content}...\nMetadata: {doc.metadata}\n")
```
